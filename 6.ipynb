```python
import folium

m = folium.Map(location=[37.547631, 126.942463], zoom_start=50)
m
```

```python
folium.Marker([37.547631, 126.942463]).add_to(m)
m
```

```python
folium.Marker([37.547631, 126.96458],popup="하트1",icon=folium.Icon(color="red",icon = "fa-solid fa-marker",prefix = "fa-solid")).add_to(m)
m
```

```python
m.add_child(folium.ClickForMarker(popup="내가클릭한곳")
m
```

```python
m.add_child(folium.LatLngPopup()) # 찍은 곳의 위도를 체크가능.
```

```python
folium.CircleMarker([
    37.547631, 126.942463],
    redius=1000,
    color='#ff0000',
    fill_color="00ff00",
    popup ="circleMarker"
    ).add_to(m)
folium.Circle([
    37.547631, 126.942463],
    redius=1000,
    color='#0000ff',
    fill_color="ffffff",
    popup ="circleMarker"
    ).add_to(m)    
m
```
