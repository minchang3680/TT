```python
import numpy as np
x = np.linspace(0, 10, 100)
y = np.sin(x)

# 플롯 생성
plt.plot(x, y)

# 주요 및 보조 눈금 활성화 및 스타일링
plt.grid(which='major', color='blue', linestyle='-', linewidth=0.8)  # 주요 눈금선
plt.minorticks_on()  # 보조 눈금 활성화
plt.grid(which='minor', color='gray', linestyle=':', linewidth=0.5)  # 보조 눈금선

# 플롯 표시
plt.show()
```

```python
import matplotlib.pyplot as plt

# 데이터 생성
labels = ['A', 'B', 'C', 'D']
sizes = [30, 25, 20, 15]  # 각 항목의 비율
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']  # 부드러운 색상
explode = (0.1, 0, 0, 0)  # 강조를 위한 Apples만 분리

# 파이차트 생성
plt.figure(figsize=(8, 8))  # 차트 크기 조정
plt.pie(
    sizes, 
    labels=labels, 
    colors=colors, 
    explode=explode, 
    autopct='%1.1f%%',  # 퍼센티지 표시
    startangle=90,      # 차트 시작 각도
    shadow=True         # 그림자 효과
)

# 제목 추가
# plt.title('Fruit Distribution', fontsize=16)

# 차트 표시
plt.show()
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales_2019 = [100,120,140,110,130,150,160,170,180,200,190,210]
sales_2020 = [90,110,130,120,140,160,170,160,150,180,200,190]

# line1 = plt.plot([1,2,3], label = "label1")
plt.plot(months, sales_2019,label = "2019")
plt.plot(months, sales_2020,label = "2020")
plt.title("Practice 1")
plt.xlabel("Month")
plt.ylabel("Sales")
# plt.xlabel("x", labelpad=30, loc="left", fontdict = font1)
# plt.ylabel("y", loc="top")
plt.legend(loc="best", ncol=2, fontsize=10, frameon=True, shadow=True)
```

```python
import matplotlib.pyplot as plt

# 데이터 생성
labels = ['Apple', 'Banana', 'Mango', 'Blueberry']
sizes = [15, 30, 45, 10]  # 각 항목의 비율
colors = ['blue', 'orange', 'green', 'red']  # 부드러운 색상
explode = (0, 0.1, 0, 0)  # 강조를 위한 Banana만 분리

# 파이차트 생성
plt.figure(figsize=(8, 8))  # 차트 크기 조정
plt.pie(
    sizes, 
    labels=labels, 
    colors=colors, 
    explode=explode, 
    autopct='%1.1f%%',  # 퍼센티지 표시
    startangle=90,      # 차트 시작 각도
    shadow=True         # 그림자 효과
)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y_sin = np.sin(x)
y_cos = np.cos(x) 

plt.plot(x, y_sin, label='sin(x)', color='blue', linestyle='-', linewidth=2)
plt.plot(x, y_cos, label='cos(x)', color='red', linestyle='-', linewidth=2)

plt.title('Sine and Cosine Functions')
plt.xlabel('x')       
plt.ylabel('y')                

plt.ylim(-1, 1)
plt.xlim(0, 10)

xticks = [0, 2, 4, 6 ,8, 10]
plt.xticks(ticks=xticks)
plt.legend()
plt.grid(which='major', color='gray', linestyle='-', linewidth=0.8) 

plt.show()
```
