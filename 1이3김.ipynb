```python
import matplotlib.pyplot as plt
from matplotlib import rc

# 한글 폰트 설정 (Windows: Malgun Gothic, Mac: AppleGothic, Ubuntu: NanumGothic)
rc('font', family='Malgun Gothic')  # Windows의 경우
# rc('font', family='AppleGothic')  # Mac의 경우
# rc('font', family='NanumGothic')  # Ubuntu의 경우

# 데이터
categories = ["스스로 그만둔 적이 있다", "해고를 당한 적이 있다", "둘 다 경험한 적이 없다"]
values = [35.3, 3, 59.5]
explode = (0, 0, 0.05)  # '둘 다 경험한 적이 없다' 항목만 강조

# 스스로 그만두거나 해고당한 비율과 둘 다 경험한 적 없다 비율 합산
combined_percentage = values[0] + values[1]

# 파이 차트 생성
plt.figure(figsize=(8, 8))
plt.pie(
    values,
    labels=categories,
    autopct='%1.1f%%',  # 퍼센트 표시
    startangle=90,      # 시작 각도
    colors=plt.cm.Paired.colors,  # 색상 팔레트
    shadow=True,       # 그림자 설정
    explode=explode,    # explode 인자 추가
    textprops={'fontsize': 17}
)

# 합산된 비율을 차트 아래에 텍스트로 추가
plt.gca().set_aspect('equal')  # 차트를 원형으로 유지
plt.figtext(0.5, -0.1, f"스스로 그만 두거나 해고당한 비율: {combined_percentage:.1f}%", 
            ha='center', fontsize=15, color='black', weight='bold')

plt.title('2020 퇴사 경험 여부 설문 결과', fontsize=18)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

# 시스템에 기본으로 제공되는 한글 폰트 설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic'  # Windows에서 한글 폰트로 Malgun Gothic을 설정
# MacOS에서는 'AppleGothic'을 기본으로 사용할 수 있습니다

# 데이터 생성
data = {
    "퇴사 이유": [
        "직무가 적성에 맞지 않음", "낮은 연봉(급여)", "조직문화불만족", 
        "높은 근무강도", "개인 역량의 발전이 어려움", "회사 비전, 성장성에 대한 의문", 
        "복리후생 불만족"
    ],
    "비율": [45.9, 36.2, 31.5, 21.4, 20.3, 18.4, 17.4]
}

# pandas DataFrame으로 변환
df = pd.DataFrame(data)

# 비율을 기준으로 내림차순 정렬
df_sorted = df.sort_values(by='비율', ascending=False)

# 막대 그래프 그리기 (세로 막대 그래프)
plt.figure(figsize=(11, 6))
bars = plt.bar(df_sorted['퇴사 이유'], df_sorted['비율'], color='skyblue')

# 그래프 제목 및 라벨
plt.ylabel('비율 (%)',fontsize=13)
plt.title('2022 기업이 생각하는 조기퇴사사유', fontsize=20)
plt.xticks(rotation=30, ha='right',fontsize = 13)  # x축 라벨 각도 조정
plt.yticks(fontsize=13)
plt.grid(True, axis='y', linestyle='--', alpha=0.7)

# 그래프 안에 값 레이블 추가
for bar in bars:
    plt.text(
        bar.get_x() + bar.get_width() / 2,  # 막대의 중앙 x 좌표
        bar.get_height() - 5,              # 막대 내부에 표시 (막대 높이보다 약간 낮게)
        f'{bar.get_height()}%',            # 표시할 텍스트
        ha='center', va='center', color='blue', fontsize=15  # 텍스트 정렬 및 색상
    )

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# 데이터 (30%만 표시)
values = [84.7, 15.3]
labels = ['84.7%', '']  # 70%는 라벨을 비워서 숨깁니다.
colors = ['#ff9999', 'white']  # 70%는 흰색으로 설정하여 숨깁니다.

# 파이 차트 그리기
plt.figure(figsize=(8, 8))
plt.pie(
    values,
    labels=labels,
    autopct='%1.1f%%',  # 퍼센트 표시
    startangle=220,      # 시작 각도
    colors=colors,      # 색상 설정
    explode=(0, 0.1),   # 30% 부분 강조
    shadow=True,        # 그림자 효과
    textprops={'fontsize': 15}
)

# 제목 설정
plt.title('1년이내 퇴사비율', fontsize=18)

# 차트 출력
plt.show()
```
